```python
"""

Este script em Python é responsável por realizar o processo de Machine Learning na base de dados Census do governo dos Estados Unidos. Ele utiliza os modelos Random Forest e Regressão Logística para análise dos dados e previsão de resultados.

Autores
Gabriely Camile de Oliveira Martins – RA: 823155179
Camilo Caetano Pinheiro de Oliveira – RA: 82212184
Eliabe Trajano de Araújo – RA: 125111372575
Rafael Henrique Koppe – RA: 202013205
Última atualização
13/06/2023

Descrição
Este script carrega os dados da base de dados Census, realiza a análise exploratória dos dados e treina modelos de Machine Learning para prever a variável alvo.

Requisitos
Python 3.10 ou superior
Bibliotecas necessárias:
Pandas
Numpy
Matplotlib
Seaborn
Sklearn
Ipywidgets
Ipython
Configuração
Antes de executar o script, é necessário configurar os seguintes parâmetros:

file_path: Caminho do arquivo CSV que contém os dados da base de dados Census do governo dos Estados Unidos.
Após a configuração, o script está pronto para ser executado.

Funcionalidades
O script oferece as seguintes funcionalidades:

Carregamento dos dados: Os dados são carregados a partir do arquivo CSV especificado.
Tratamento de dados ausentes: É possível selecionar a estratégia de tratamento de dados ausentes, incluindo a remoção de linhas com valores ausentes ou o preenchimento desses valores com estratégias como média, mediana, valor mais frequente ou valor constante.
Seleção da variável alvo: É possível selecionar a variável alvo para análise e previsão.
Definição da quantidade de dados de treino: É possível definir a porcentagem de dados a serem utilizados para treinamento dos modelos. O restante dos dados será usado para teste.
Análise dos dados: O script realiza uma análise exploratória dos dados, exibindo estatísticas gerais, estatísticas da variável alvo e visualizações como matriz de correlação e importância das variáveis.
Treinamento e avaliação dos modelos: O script treina os modelos de Random Forest e Regressão Logística com os dados de treino e avalia sua acurácia usando os dados de teste. Também exibe as matrizes de confusão dos modelos.
Visualizações: O script exibe visualizações como a matriz de correlação e a importância das variáveis.
Utilização
Para utilizar o script, siga as etapas a seguir:

Configure os parâmetros necessários, como o caminho do arquivo CSV e as opções de tratamento de dados ausentes.
Execute o script em um ambiente Python com as bibliotecas necessárias instaladas.
No ambiente interativo, selecione a variável alvo e a estratégia de tratamento de dados ausentes.
Informe a porcentagem de dados de treino desejada.
Clique no botão "Analyze" para iniciar a análise dos dados e o treinamento dos modelos.
Os resultados serão exibidos na saída do script, incluindo estatísticas, matrizes de confusão e visualizações.

Link de acesso à documentação completa: https://docs.google.com/document/d/1rCkDEI4LI8afra2vGKukqMu8IvGEs2JF/edit

"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import warnings
import ipywidgets as widgets
from IPython.display import display
import gdown
from tabulate import tabulate

# Load data from the CSV file


# URL do link de compartilhamento do Google Drive
url = 'https://drive.google.com/uc?id=1zN2FLFTboKAD7eu_9Xh1PCVbAb18jCas'

# Faz o download do arquivo usando o gdown
output = 'DadosDoCensus.csv'
gdown.download(url, output, quiet=False)

# Lê o arquivo CSV com o Pandas
df = pd.read_csv(output)

#file_path = '/content/drive/MyDrive/Colab Notebooks/Dados.csv'

#try:
#    df = pd.read_csv(file_path)
#except FileNotFoundError:
#    raise FileNotFoundError(f"File not found: {file_path}")

print()
print('------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
# Limitar a exibição a 10 linhas
df_limitado = df.head(10)

# Exibir o DataFrame limitado como uma tabela formatada
print(df_limitado.to_string(index=False))
print('------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')

# Function to determine if the variable is binary
def is_binary_variable(column):
    unique_values = column.unique()
    return len(unique_values) == 2 and all(value in [0, 1] for value in unique_values)

class CustomLogisticRegression(LogisticRegression):
    def fit(self, X, y):
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=UserWarning)
            super().fit(X, y)

class CustomRandomForestClassifier(RandomForestClassifier):
    def fit(self, X, y):
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=UserWarning)
            super().fit(X, y)

# Select the target variable
target_variable = widgets.Dropdown(options=df.columns.tolist(), description='Target Variable:')
display(target_variable)

# Select the missing data handling strategy
imputer_strategy = widgets.Dropdown(
    options=['mean', 'median', 'most_frequent', 'constant', 'remove'],
    description='Missing Data Handling Strategy:'
)
display(imputer_strategy)

# Input field for entering the percentage of training data
train_data_percentage = widgets.FloatText(value=70, description='Training Data Percentage (%):')
display(train_data_percentage)

def analyze_button_clicked(b):
    # Display the target variable
    print(f"Target Variable: {target_variable.value}")

    def analyze_data():
        # Define X and y based on the selected target variable
        X = df.drop(columns=[target_variable.value])
        y = df[target_variable.value]

        # General statistics for the dataset
        print("Dataset Statistics:")
        display(df.describe(include='all'))
        print()

        # Statistics for the target variable
        print("Target Variable Statistics:")
        display(y.value_counts())
        print()


        # Split the data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=1 - train_data_percentage.value / 100, random_state=42
        )

        # Update the number of training and testing rows
        training_rows = X_train.shape[0]
        testing_rows = X_test.shape[0]
        print(f"Number of Training Rows: {training_rows}")
        print(f"Number of Testing Rows: {testing_rows}")
        print()

        if imputer_strategy.value == 'remove':
            # Remove rows with missing values
            original_shape = X_train.shape[0]
            X_train = X_train.dropna()
            y_train = y_train[X_train.index]
            X_test = X_test.dropna()
            y_test = y_test[X_test.index]
            removed_rows = original_shape - X_train.shape[0]
            print(f"Removed {removed_rows} rows with missing values.")
            print(f"Total rows: {original_shape}")
            print(f"Remaining rows: {X_train.shape[0]}")
            print(f"Impact on analysis: {(removed_rows / original_shape) * 100}%")
            print()
            # Update the number of training and testing rows
            training_rows = X_train.shape[0]
            testing_rows = X_test.shape[0]
            print(f"Number of Training Rows: {training_rows}")
            print(f"Number of Testing Rows: {testing_rows}")
            print()

        else:
            # Fill missing values using an imputer
            if imputer_strategy.value == 'mean':
                imputer = SimpleImputer(strategy='mean')
            elif imputer_strategy.value == 'median':
                imputer = SimpleImputer(strategy='median')
            elif imputer_strategy.value == 'most_frequent':
                imputer = SimpleImputer(strategy='most_frequent')
            elif imputer_strategy.value == 'constant':
                imputer = SimpleImputer(strategy='constant', fill_value=0)  # Replace 0 with the desired constant value
            X_train = imputer.fit_transform(X_train)
            X_test = imputer.transform(X_test)

        # Standardize the data
        scaler = StandardScaler()
        X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
        X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)

        logistic_regression_model = CustomLogisticRegression()
        random_forest_model = CustomRandomForestClassifier(n_estimators=100, random_state=42)

        # Function to train and evaluate the models
        def train_and_evaluate_model(model, X_train, X_test, y_train, y_test):
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)
            cm = confusion_matrix(y_test, y_pred)
            return accuracy, cm

        # Train and evaluate the models
        logistic_regression_accuracy, logistic_regression_cm = train_and_evaluate_model(
            logistic_regression_model, X_train, X_test, y_train, y_test
        )
        random_forest_accuracy, random_forest_cm = train_and_evaluate_model(
            random_forest_model, X_train, X_test, y_train, y_test
        )

        # Display the results
        print("Logistic Regression Accuracy:", logistic_regression_accuracy)
        print("Logistic Regression Confusion Matrix:")
        plt.figure()
        sns.heatmap(logistic_regression_cm, annot=True, fmt='d', cmap='Blues')
        plt.title("Logistic Regression Confusion Matrix")
        plt.show()
        print()
        print("Random Forest Accuracy:", random_forest_accuracy)
        print("Random Forest Confusion Matrix:")
        plt.figure()
        sns.heatmap(random_forest_cm, annot=True, fmt='d', cmap='Blues')
        plt.title("Random Forest Confusion Matrix")
        plt.show()
        print()

        # Display the variable importance plot
        random_forest_model.fit(X_train, y_train)
        importances = random_forest_model.feature_importances_
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(10, 6))
        plt.bar(range(X_train.shape[1]), importances[indices])
        plt.xticks(range(X_train.shape[1]), X_train.columns[indices], rotation=90)
        plt.title("Variable Importance")
        plt.tight_layout()
        plt.show()

        # Display the correlation matrix plot
        correlation_matrix = X_train.corr()
        plt.figure(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
        plt.title("Correlation Matrix")
        plt.tight_layout()
        plt.show()

    analyze_data()

analyze_button = widgets.Button(description="Analyze")
analyze_button.on_click(analyze_button_clicked)
display(analyze_button)
```
